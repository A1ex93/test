```python
import pandas as pd
import numpy as np
from tensorflow import keras
import matplotlib.pyplot as plt

# 1. Загрузка и подготовка данных
# ---------------------------------------------------
# Загрузка данных (одна строка = метка + 784 пикселя через пробел)
df = pd.read_csv('mnist_test.csv', header=None, sep=',')

# Разделение на метки (первый элемент) и пиксели (остальные 784)
y = df.iloc[:, 0].values  # Все метки
X = df.iloc[:, 1:785].values  # Все пиксели (возьмем ровно 784 столбца)

# Преобразование в изображения 28x28
X = X.reshape(-1, 28, 28)

print(f"Всего загружено {len(X)} изображений")
print("Уникальные метки:", np.unique(y))

# 2. Фильтрация данных (выбираем только цифры 1, 3, 5)
# ---------------------------------------------------
digits = [8, 5, 1]  # Какие цифры оставляем

# Создаем маску для фильтрации
mask = np.isin(y, digits)
X_filtered = X[mask]
y_filtered = y[mask]

print(f"\nПосле фильтрации осталось {len(X_filtered)} изображений")

# 3. Разделение на обучающую и тестовую выборки
# ---------------------------------------------------
X_train, X_test, y_train, y_test = [], [], [], []

for digit in digits:
    # Находим индексы для текущей цифры
    indices = np.where(y_filtered == digit)[0]

    # Берем первые 20 для обучения, следующие 10 для теста
    X_train.append(X_filtered[indices[:20]])
    X_test.append(X_filtered[indices[20:30]])
    y_train.extend([digit] * 20)
    y_test.extend([digit] * 10)

# Объединяем и перемешиваем данные
X_train = np.concatenate(X_train)
X_test = np.concatenate(X_test)
y_train = np.array(y_train)
y_test = np.array(y_test)

# Перемешиваем
shuffle_train = np.random.permutation(len(X_train))
shuffle_test = np.random.permutation(len(X_test))
X_train, y_train = X_train[shuffle_train], y_train[shuffle_train]
X_test, y_test = X_test[shuffle_test], y_test[shuffle_test]

# Преобразуем метки в one-hot encoding
from sklearn.preprocessing import LabelEncoder

# Создаем кодировщик для перевода [8,5,1] → [0,1,2]
label_encoder = LabelEncoder()
label_encoder.fit(digits)  # Обучаем на доступных цифрах

# Перекодируем метки
y_train_encoded = label_encoder.transform(y_train)
y_test_encoded = label_encoder.transform(y_test)

# Теперь можно безопасно применить to_categorical
y_train = keras.utils.to_categorical(y_train_encoded, num_classes=len(digits))
y_test = keras.utils.to_categorical(y_test_encoded, num_classes=len(digits))
# Нормализуем пиксели
X_train = X_train / 255.0
X_test = X_test / 255.0

print("\nИтоговые размеры выборок:")
print(f"Обучающая: X={X_train.shape}, y={y_train.shape}")
print(f"Тестовая: X={X_test.shape}, y={y_test.shape}")

# 4. Создание и обучение модели
# ---------------------------------------------------
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(len(digits), activation='softmax')
])

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Обучение на 5 эпохах
print("\nОбучение на 5 эпохах:")
history_5 = model.fit(X_train, y_train, epochs=5, validation_data=(X_test, y_test))

# Обучение на 10 эпохах
print("\nОбучение на 10 эпохах:")
history_10 = model.fit(X_train, y_train, epochs=10, validation_data=(X_test, y_test))

# 5. Оценка результатов
# ---------------------------------------------------
def print_accuracy(history, epochs):
    train_acc = history.history['accuracy'][-1] * 100
    test_acc = history.history['val_accuracy'][-1] * 100
    print(f"Точность после {epochs} эпох:")
    print(f"- Обучающая выборка: {train_acc:.2f}%")
    print(f"- Тестовая выборка: {test_acc:.2f}%")

print_accuracy(history_5, 5)
print_accuracy(history_10, 10)

# Визуализация нескольких примеров
plt.figure(figsize=(10, 5))
for i in range(5):
    plt.subplot(1, 5, i+1)
    plt.imshow(X_train[i], cmap='gray')
    plt.title(f"Метка: {np.argmax(y_train[i])}")
    plt.axis('off')
plt.show()
```
